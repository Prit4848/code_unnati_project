```python
import matplotlib.pyplot as plt

years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
net_income = [631, 614, 1666, 3047, 4141, 2796, 4332, 9752, 4368, 29760]

plt.figure(figsize=(10, 6))
plt.plot(years, net_income, marker='o', color='b', linewidth=2, markersize=8)
plt.title('Year vs. Net Income', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Net Income (mil. US$)', fontsize=14)
plt.xticks(years)
plt.grid(True)
plt.tight_layout()
plt.show()
```
